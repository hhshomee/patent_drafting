```python
import pandas as pd
df=pd.read_csv('g_patent.tsv', sep='\t')
filtered_patent_df = pd.read_csv('filtered_patents.csv')
merged_df = pd.merge(filtered_patent_df, df, on='patent_id', how='inner')
aggregated_df = merged_df.groupby('patent_id').agg({
    'patent_abstract': 'first',
    'patent_title': 'first'
}).reset_index()
aggregated_df.to_csv('aggregated_patent_data_H04.csv', index=False)
```

```python
import pandas as pd
df=pd.read_csv('g_claims_2022.tsv', sep='\t')
aggregated_df = pd.read_csv('aggregated_patent_data.csv')
merged_df = pd.merge(aggregated_df, df, on='patent_id', how='inner')
aggregated_df = merged_df.groupby('patent_id').agg({
    'patent_abstract': 'first',
    'patent_title': 'first',
    'claim_text': 'first'
}).reset_index()
aggregated_df.to_csv('aggregated_patent_data.csv', index=False)
```
