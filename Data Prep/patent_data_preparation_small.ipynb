```python
'''This is for subclass level
we need only 'patent_id' and 'cpc_subclass' columns'''

import pandas as pd
filtered_df_G06 = df_cpc_current[df_cpc_current['cpc_subclass'] == 'G06V']
selected_columns = filtered_df_G06[['patent_id', 'cpc_subclass']]
selected_columns.to_csv('filtered_patents_G06V.csv', index=False)
```

```python
import pandas as pd

filtered_patent_df = pd.read_csv('filtered_patents_G06V.csv')
merged_df = pd.merge(filtered_patent_df, df_g_patent, on='patent_id', how='inner')
new_df = merged_df[['patent_id', 'patent_abstract', 'patent_title','cpc_subclass']]
new_df.to_csv('aggregated_patent_data_G06V.csv', index=False)
```

```python
aggregated_patent_df = pd.read_csv('/content/aggregated_patent_data_G06V.csv')
```

```python
aggregated_patent_df.head()
```

```python
aggregated_patent_df.shape
```

```python
import pandas as pd

'''We are merging  the two df (aggregated and claims) based on 'patent_id
after that group by patent id to get our desired field: abstract, title, claim and cpc code'''

merged_df = pd.merge(aggregated_patent_df, df_g_claims, on='patent_id', how='inner')


aggregated_df = merged_df.groupby('patent_id').agg({
    'patent_abstract': 'first',
    'patent_title': 'first',
    'claim_text': 'first',
    'cpc_subclass':'first'

}).reset_index()
aggregated_df.to_csv('patent_data_G06V.csv', index=False)
```
