```python
from rouge import Rouge

def calculate_rouge(hypotheses, references):
    """
    Calculate ROUGE scores for a set of hypotheses and reference sentences.

    Args:
        hypotheses: a list of strings (each a generated text)
        references: a list of strings (each a reference text)

    Returns:
        Dictionary with ROUGE-1 and ROUGE-2 scores
    """

    rouge = Rouge()
    scores = rouge.get_scores(hypotheses, references, avg=True)

    return scores


hypotheses = df['generated_abstract'].astype(str).tolist()
references = df['patent_abstract'].astype(str).tolist()
rouge_scores = calculate_rouge(hypotheses, references)
r=[]
p=[]
f=[]
for i in range(len(references)):
  rouge_scores = calculate_rouge(hypotheses[i], references[i])
  dictionary =rouge_scores['rouge-l']
  R=dictionary['r']
  r.append(R)
  P=dictionary['p']
  p.append(P)
  F=dictionary['f']
  f.append(F)
df.loc[:, 'rouge_Precision'] = [float(f"{score:.2f}") for score in p]
df.loc[:, 'rouge_Recall'] = [float(f"{score:.2f}") for score in r]
df.loc[:, 'rouge_F1_Score'] = [float(f"{score:.2f}") for score in f]

df.to_csv('Evaluation_abstract_A61B_rouge.csv', index=False)
```
