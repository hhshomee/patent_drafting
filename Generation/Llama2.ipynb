```python
from huggingface_hub import hf_hub_download
model_path = hf_hub_download(repo_id=model_name, filename=model_basename)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from llama_index.core import (
    VectorStoreIndex,
    ServiceContext,
)
from llama_index.llms.llama_cpp import LlamaCPP
from llama_index.llms.llama_cpp.llama_utils import messages_to_prompt
```

```python
llm = LlamaCPP(
    model_url=None,
    model_path=model_path,
    temperature=0.1,
    max_new_tokens=512,
    context_window=3000,
    generate_kwargs={},
    model_kwargs={"n_gpu_layers": 43},
    messages_to_prompt=messages_to_prompt,
    verbose=True,
)
```

```python
import pandas as pd
df = pd.read_csv('/content/patent_data_A61B.csv')
```

```python
df.head()
```

```python
df['generated_abstract'] = pd.NA
def generate_abstract(claim_text):
    try:
        prompt = f"Given the patent claim: {claim_text}\ngenerate the abstract."
        response_iter = llm.stream_complete(prompt)
        abstract = "".join([response.delta for response in response_iter])
        return abstract
    except Exception as e:
        print("An error occurred:", e)
        return None

df.iloc[0:0].to_csv('GeneratedAbstract_A61B.csv', index=False)

for index, row in df.iterrows():
    print(index)
    generated_abstract = generate_abstract(row['claim_text'])
    df.at[index, 'generated_abstract'] = generated_abstract if isinstance(generated_abstract, str) else 'Error or empty'
    df.iloc[[df.index.get_loc(index)]].to_csv('GeneratedAbstract_A61B.csv', mode='a', header=False, index=False)
```
