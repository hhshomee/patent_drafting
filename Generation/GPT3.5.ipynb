```python
import os
from openai import OpenAI
```

```python
import pandas as pd
df['generated_abstract'] = pd.NA
def generate_abstract(claim_text):
    client = OpenAI(
        api_key="###############". #insert your API key from OpenAI
    )

    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "assistant",
                "content": f"Given the patent claim write the abstract. Claim: {claim_text}",
            }
        ],
        model="gpt-3.5-turbo",
    )
    abstract = chat_completion.choices[0].message.content
    return abstract
df.iloc[0:0].to_csv('GPT3.5_A61B.csv', index=False)

for index, row in df.iterrows():
    print(index)

    generated_abstract = generate_abstract(row['claim_text'])

    df.at[index, 'generated_abstract'] = generated_abstract
    df.iloc[[index]].to_csv('GPT3.5_A61B.csv', mode='a', header=False, index=False)
```
